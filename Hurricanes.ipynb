```python
#Damages function here
def update_damages(damages_list):
    updated_damages = []
    for damage in damages_list:
        if "M" in damage:
            updated_damages.append(float(damage.strip("M")) * 1000000)
        elif "B" in damage:
            updated_damages.append(float(damage.strip("B")) * 1000000000)
        else:
            updated_damages.append("Damages not recorded")
    return updated_damages

#Test the function
#print(update_damages(damages)[13])
```

```python
#Construct hurricanes dictionary function
def hurricanes_dict(names, months, years, max_sustained_winds, areas_affected, damages, deaths):
    hurricanes_dict = {}
    for i in range(len(names)):
        hurricanes_dict.update({names[i]: {"Name": names[i], "Month": months[i], "Year": years[i], \
        "Max Sustained Wind": max_sustained_winds[i], "Areas Affected": areas_affected[i], "Damage": damages[i], \
        "Death": deaths[i]}})
    return hurricanes_dict
hurricanes = hurricanes_dict(names, months, years, max_sustained_winds, areas_affected, update_damages(damages), deaths)
```

```python
#Count affected areas function
def affected_areas(dictionary):
    affected_areas = {}
    unique_areas = []
    for hurricane in dictionary.values():
        for area in hurricane["Areas Affected"]:
            if area not in unique_areas:
                unique_areas.append(area)
    for area in unique_areas:
        counter = 0
        for hurricane in dictionary.values():
            for area2 in hurricane["Areas Affected"]:
                if area == area2:
                    counter += 1
        affected_areas.update({area: counter})
    return affected_areas

affected_areas_count = affected_areas(hurricanes)
#Function test
#print(affected_areas(hurricanes))
```

```python
#catgeorize by mortality function

mortality_scale = {0: 0,
                   1: 100,
                   2: 500,
                   3: 1000,
                   4: 10000,
                   5: 10001}

#Whent a little futher on the exercise and created a function with the mortality scale as an input
def by_mortality(hurricanes, mortality_scale):
    hurricanes_by_mortality = {}
    for rating in mortality_scale:
        temp_list= []
        for hurri in hurricanes:
            if hurricanes[hurri]["Death"] <= mortality_scale[rating] and hurricanes[hurri]["Death"] > mortality_scale.get(rating - 1, -1):
                temp_list.append(hurricanes[hurri])
            elif rating == len(mortality_scale) -1 and hurricanes[hurri]["Death"] >= mortality_scale[rating]:
                temp_list.append(hurricanes[hurri])
        hurricanes_by_mortality.update({rating: temp_list})
    return hurricanes_by_mortality

by_mortality = by_mortality(hurricanes, mortality_scale)
#Function test
#print(by_mortality)
```

```python
#catgeorize by damage function
damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000,
                5: 50000000001}

def hurricanes_by_damage_scale(hurricanes, damage_scale):
    hurricanes_by_scale = {}
    for rating in damage_scale:
        temp_list = []
        for hurri in hurricanes:
            if not hurricanes[hurri]["Damage"] == "Damages not recorded":
                if hurricanes[hurri]["Damage"] > damage_scale.get(rating - 1, -1) and hurricanes[hurri]["Damage"] <= damage_scale[rating]:
                    temp_list.append(hurricanes[hurri])
                elif rating == len(damage_scale) -1 and hurricanes[hurri]["Damage"] >= damage_scale[rating]:
                    temp_list.append(hurricanes[hurri])
        hurricanes_by_scale.update({rating: temp_list})
    return hurricanes_by_scale

hurricanes_damage_scale = hurricanes_by_damage_scale(hurricanes, damage_scale)
#Function test
#print(hurricanes_damage_scale)
```
