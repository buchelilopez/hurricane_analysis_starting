```python
#Damages function here
def update_damages(damages_list):
    updated_damages = []
    for damage in damages_list:
        if "M" in damage:
            updated_damages.append(float(damage.strip("M")) * 1000000)
        elif "B" in damage:
            updated_damages.append(float(damage.strip("B")) * 1000000000)
        else:
            updated_damages.append("Damages not recorded")
    return updated_damages

#Test the function
#print(update_damages(damages)[13])
```

```python
#Construct hurricanes dictionary function
def hurricanes_dict(names, months, years, max_sustained_winds, areas_affected, damages, deaths):
    hurricanes_dict = {}
    for i in range(len(names)):
        hurricanes_dict.update({names[i]: {"Name": names[i], "Month": months[i], "Year": years[i], \
        "Max Sustained Wind": max_sustained_winds[i], "Areas Affected": areas_affected[i], "Damage": damages[i], \
        "Death": deaths[i]}})
    return hurricanes_dict
#print(hurricanes_dict(names, months, years, max_sustained_winds, areas_affected, update_damages(damages), deaths))
```

```python
#hurricane by year dictionary function
def hurricanes_by_year(dictionary):
    hurracanes_years = {}
    for hurricane in dictionary.values():
        temp_list = []
        for hurricane2 in dictionary.values():
            if hurricane["Year"] == hurricane2["Year"]:
                temp_list.append(hurricane2)
        hurracanes_years.update({hurricane["Year"]: temp_list})
    return hurracanes_years

#Function test
#dictionary_per_year = hurricanes_by_year(hurricanes_dict(names, months, years, max_sustained_winds, areas_affected, update_damages(damages), deaths))
#print(dictionary_per_year)
#for key, value in dictionary_per_year.items():
    #print(key, ":", len(value))
```
